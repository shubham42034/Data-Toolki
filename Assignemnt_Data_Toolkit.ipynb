6.  What are the differences between NumPy arrays and Python lists

->

#. Data Type
NumPy Arrays: Homogeneous; all elements must be of the same data type.
Python Lists: Heterogeneous; elements can be of different data types.
#. Performance
NumPy Arrays: Faster due to optimized C-based implementation and contiguous memory allocation.
Python Lists: Slower as they store references to objects, leading to overhead.
#. Memory Efficiency
NumPy Arrays: More memory-efficient because of fixed data types and compact storage.
Python Lists: Less efficient as they store pointers to objects.
#. Operations
NumPy Arrays: Support element-wise operations and advanced mathematical computations directly.
Python Lists: Require explicit loops or list comprehensions for similar operations.
#. Flexibility
NumPy Arrays: Less flexible; resizing or appending is not straightforward.
Python Lists: More flexible; can dynamically grow or shrink.
#. Use Case
NumPy Arrays: Ideal for numerical and scientific computing.
Python Lists: Better for general-purpose programming.
In summary, NumPy arrays are optimized for numerical computations and large datasets, while Python lists are more versatile for everyday programming tasks.

17.  Explain the difference between apply() and map() in Pandas

->  n Pandas, both apply() and map() are used to apply functions to data, but they differ in their scope and usage:

1. map()
Scope: Works only on Pandas Series (i.e., a single column or row).
Functionality: Applies a function element-wise to each value in the Series.
Use Case: Ideal for transforming or mapping values in a single column or row.
Example:

import pandas as pd

# Example Series
s = pd.Series([1, 2, 3, 4])

# Using map to square each element
result = s.map(lambda x: x**2)
print(result)
Output:

0     1
1     4
2     9
3    16
dtype: int64
2. apply()
Scope: Works on both Pandas Series and DataFrames.

Functionality:

For a Series, it applies a function element-wise (similar to map()).
For a DataFrame, it applies a function along an axis (rows or columns).
Use Case: More versatile; used for row-wise or column-wise operations in DataFrames or element-wise operations in Series.

Example (on a Series):


# Using apply to square each element in a Series
result = s.apply(lambda x: x**2)
print(result)
Output (same as map()):


0     1
1     4
2     9
3    16
dtype: int64
Example (on a DataFrame):


# Example DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# Using apply to sum values row-wise
result = df.apply(lambda row: row.sum(), axis=1)
print(result)
Output:


0     5
1     7
2     9
dtype: int64
Key Differences
Feature	map()	apply()
Scope	Series only	Series and DataFrames
Functionality	Element-wise operations	Element-wise (Series) or axis-wise (DataFrame)
Flexibility	Limited to Series transformations	More versatile for complex operations
In summary, use map() for simple element-wise transformations on Series, and apply() for more complex operations, especially when working with DataFrames.

21.   Why is NumPy’s array slicing faster than Python’s list slicing


->NumPy's array slicing is faster than Python's list slicing due to several key reasons:

Contiguous Memory Storage:
NumPy arrays store data in contiguous memory blocks, which allows for efficient memory access. Python lists, on the other hand, store elements in non-contiguous memory locations, leading to slower access times.

Homogeneous Data Types:
NumPy arrays are homogeneous, meaning all elements are of the same data type. This allows NumPy to perform operations directly at the hardware level using optimized C implementations. Python lists, being heterogeneous, require additional overhead to handle different data types.

Vectorized Operations:
NumPy leverages vectorized operations, which means slicing and other operations are executed in bulk at the low level, avoiding Python's slower interpreted loops. Python lists rely on Python's slower iteration mechanisms.

No Type Checking Overhead:
Since NumPy arrays are homogeneous, there is no need for type checking during slicing. Python lists, however, require type checks for each element, adding to the overhead.

Efficient Indexing:
NumPy arrays allow direct indexing into the memory block, making slicing operations almost instantaneous. Python lists involve additional steps to locate elements due to their non-contiguous structure.

In summary, NumPy's design, optimized for numerical computations, ensures faster slicing and overall performance compared to Python lists.
